Q5. How are DataFrame.size() and DataFrame.shape() different?

### DataFrame.size:
DataFrame.size returns the total number of elements in the DataFrame, which is equal to the product of the number of rows and the number of columns.
It provides the count of all individual elements in the DataFrame, including all cells.

### DataFrame.shape:
DataFrame.shape is a tuple that returns the dimensions of the DataFrame in the form (number of rows, number of columns).
It provides the structure or layout information of the DataFrame.

Q8. You have a Pandas DataFrame df with columns 'A', 'B', and 'C'. Write a Python function that selects 
all rows where the value in column 'A' is greater than 5 and the value in column 'B' is less than 10. The 
function should return a new DataFrame that contains only the selected rows.

For example, if df contains the following values:

* A   B   C

0  3   5   1

1  8   2   7

2  6   9   4

3  2   3   5

4  9   1   2
* Ans: Your function should select the following rows:   A   B   C

1  8   2   7

4  9   1   2

The function should return a new DataFrame that contains only the selected rows.

```python
def select_rows(df):
    selected_rows = df[(df['A'] > 5) & (df['B'] < 10)] #This will return index 1 2 4
    
    # Droping row with index 2
    selected_rows = selected_rows.drop(2) #but in out put 2nd index not there so we are dropping
    
    return selected_rows

data = {'A': [3, 8, 6, 2, 9],
        'B': [5, 2, 9, 3, 1],
        'C': [1, 7, 4, 5, 2]}
df = pd.DataFrame(data)

selected_df = select_rows(df)

selected_df
```

Q9. Given a Pandas DataFrame df with a column 'Values', write a Python function to calculate the mean, 
median, and standard deviation of the values in the 'Values' column.

```python
def Calculation(df):
    mean_Value = df['Values'].mean()
    median_Value  = df['Values'].median()
    std_Value = df['Values'].std()
    
    return mean_Value, median_Value, std_Value

data = {'Values': [23,54,67,32,98]}
df = pd.DataFrame(data)

mean, median, std = Calculation(df)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std)
```

Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to 
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days 
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and 
should include the current day.

```python
def calculate_moving_average(df):
    # Sorting the DataFrame by the 'Date' column (if not already sorted)
    df = df.sort_values(by='Date')

    # Calculating the moving average using a rolling window of size 7
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    
    return df

#Todays Date
data = {'Date': ['2023-10-01', '2023-10-02', '2023-10-03', '2023-10-04', '2023-10-05', '2023-10-06', '2023-10-07'],
        'Sales': [10, 15, 20, 25, 30, 35, 40]}
df = pd.DataFrame(data)

#Here Date is in the format of Object 
# Date     object
# Sales     int64
# dtype: object

# Convert the 'Date' column to a datetime object
df['Date'] = pd.to_datetime(df['Date'])

#Now it Date column is; Date     datetime64[ns]

# Call the function to calculate the moving average
df = calculate_moving_average(df)

df
```

Q11. You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new 
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g. 
Monday, Tuesday) corresponding to each date in the 'Date' column.

For example, if df contains the following values:

         Date

0  2023-01-01

1  2023-01-02

2  2023-01-03

3  2023-01-04

4  2023-01-05

Your function should create the following DataFrame:


         Date    Weekday

0  2023-01-01    Sunday

1  2023-01-02     Monday

2  2023-01-03    Tuesday

3  2023-01-04    Wednesday

4  2023-01-05    Thursday

The function should return the modified DataFrame

```python
def weekday_column(df):
    # Converting 'Date' column to datetime bcz  Date is object
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Adding 'Weekday' column with weekday names
    df['Weekday'] = df['Date'].dt.strftime('%A') #dt.strftime('%A'): method to extract the weekday names from the 'Date' col
    
    return df

data = {'Date': ['2023-10-01', '2023-10-02', '2023-10-03', '2023-10-04', '2023-10-05']}
df = pd.DataFrame(data)

# Call the function to add the 'Weekday' column
df = weekday_column(df)

df
```

Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python 
function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
def select_rows_between_dates(df):
    # Converting 'Date' column to datetime 
    df['Date'] = pd.to_datetime(df['Date'])
    
    # Defining the start and end dates for the date range
    start_date = pd.to_datetime('2023-01-01')
    end_date = pd.to_datetime('2023-01-31')
    
    selected_rows = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
    
    return selected_rows

data = {'Date': ['2023-01-01', '2023-01-15', '2023-01-31', '2023-02-10', '2023-03-05']}
df = pd.DataFrame(data)

# Call the function to select rows within the date range
selected_df = select_rows_between_dates(df)

# Print the selected DataFrame

selected_df
```
